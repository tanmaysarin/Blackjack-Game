```python
class Card:
    '''
    A Card class to define all the cards that are supposed to be in the deck.
    '''
    def __init__(self, suit, rank): # initialize the Card class
        self.suit = suit
        self.rank = rank
        
    def __str__(self): # make a string function, used to print the contents at any point
        return self.rank + ' of ' + self.suit
```

```python
class Deck:
    '''
    A Deck class to define the deck and store all the functions which can be performed by the deck.
    '''
    def __init__(self): # initialize the deck
        self.deck = [] # starting with an empty list    
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit, rank)) # Append 52 unique cards to the deck
    
    def __str__(self):
        deck_comp = ''
        for card in self.deck:
            deck_comp = deck_comp + '\n ' + card.__str__() # call the str function of the card class to get the 
                                                           # definition of each card
        return 'The deck has:' + deck_comp
    
    def shuffle(self): # shuffle the contents of the deck, i.e. the cards
        random.shuffle(self.deck)
    
    def deal(self): # return one card from the deck to the player or the dealer
        single_card = self.deck.pop()
        return single_card
```

```python
def hit(deck, hand):
    '''
    A function called when the player or the dealer takes a new card
    Input: Deck (the current state of the deck)
           Hand (The state of the hand of the current user(player/dealer))
    '''
    hand.add_card(deck.deal())
    hand.adjust_for_ace()
```

```python
def hit_or_stand(deck, hand):
    '''
    A function, which asks the player whether he wants to take a new card or not
    Input: Deck (the current state of the deck)
           Hand (The state of the hand of the current user(player/dealer)) 
    '''
    global playing # Global variable, set to True as default
    
    while True:
        x = input("Would you like to hit or stand? Enter 'h' or 's' -> ")
        
        if x[0].lower == 'h':
            hit(deck, hand)
        elif x[0].lower == 's':
            print("Player stands. Dealer is playing.")
            playing = False
        else:
            print("Sorry, please try again.")
            continue
        break
```
